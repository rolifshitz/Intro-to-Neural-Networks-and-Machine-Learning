### 1.1 - Implement random search and normalize the directions

```python
def random_search(g, alpha_choice, max_its, w, num_samples):
    """
    This function conducts random search to minimize a cost function g.
    Args:
        g: the function we are trying to minimize (cost functin)
        alpha_choice: The step length/learning rate, can take either the
            string 'diminishing' or a numerical constant.
        max_its: maximum numbe rof iterations to be taken
        w: the starting vector
        num_samples: how many initial random directions begin with
    Returns:
        w_history: the points that the random search steps from
        cost_history: the cost at each step point (??ITR)
    """
    # run random search
    w_history = []  # container for w history
    cost_history = []  # container for corresponding cost function history
    alpha = 0

    for k in range(1, max_its + 1):

        # Check if diminishing steplength rule used
        if alpha_choice == 'diminishing':
            alpha = 1 / float(k)
        else:
            alpha = alpha_choice
        cost_history.append(g(w))
        w_history.append(w)
        
        # Generate random directions
        w_size = np.size(w)
        rand_directions = np.random.randn(num_samples, w_size)

        # Normalize the directions
        norms = np.zeros((num_samples, w_size))
        for i in range(0, len(rand_directions)):
            norms[i] = np.linalg.norm(rand_directions[i])
        normalized_directions = rand_directions / norms  # Each row of normalized_directions is a direction.

        # Calculate possible weights based on direction
        possible_weights = w + alpha*normalized_directions
        
        # Check for steepest descent direction
        possible_costs = []
        for weight in possible_weights:
            possible_costs.append(g(weight))
        possible_costs = np.array(possible_costs)
    
        i = np.argmin(possible_costs)

        # Step if get lower cost
        if g(possible_weights[i]) < g(w):
            w = w + alpha*normalized_directions[i,:]

    cost_history.append(g(w))
    w_history.append(w)
    
    return w_history, cost_history
```

## Exercise 2: Curse of Dimensionality

```python
def curse_of_dimensionality(g, alpha_choice, P):
    """
    This function finds the fraction of P randomly generated directions
    that are descent directions for the cost function g, for initial
    starting vectors w0 = [1,0,...,0,0] of size N, where N = 1,2,...,25.
    Args:
        g: the cost function
        alpha_choice: The step length/learning rate; can take either the
            string 'diminishing' or a numerical constant.
        max_its (int): maximum number of iterations to be taken
        P (int): the number of random directions to be generated
    Returns:
        fraction_descending (list): list containing fraction of descending 
            directions for each set of randomly generate directions of a certain size. 
            Contains 25 entries, where each entry is a list of the form: 
            [vector size N, fraction descending for that vector size] 
    """

    # Generate new starting vectors
    n = 0
    start_weights = []
    while n <= 24:
        start_weights.append([1])
        n += 1
    for i in range(1,25):
        for j in range(0,i):
            start_weights[i].append(0)
    for i in range(0,25):
        start_weights[i] = np.array(start_weights[i])
        # start_weights is a list of numpy arrays, each array is a starting vector for random search.
        
        # Check if diminishing steplength rule used
        if alpha_choice == 'diminishing':
            alpha = 1 / float(k)
        else:
            alpha = alpha_choice

        all_rand_directions = [] # each element is a list of random directions generated for one start_weight size.
        # Generate random directions for each weight input size.
        for i in range(0, len(start_weights)):
            # Generate random directions with size 1 x i+1
            w_size = np.size(start_weights[i])
            rand_directions = np.random.randn(P, w_size)
            
            # Normalize the directions
            norms = np.zeros((P, w_size))
            for i in range(0, len(rand_directions)):
                norms[i] = np.linalg.norm(rand_directions[i])
            normalized_directions = rand_directions / norms  # Each row of normalized_directions is a direction.
            
            # Add the set (a list of lists) of normalized random directions with size 1 x i+1 to all_random_directions
            all_rand_directions.append(normalized_directions)
        
        # Initialize list of lists to keep track of the fraction of descending 
        # directions for each set of random directions of a certain size.
        # Each list within fraction descending is of the form: 
        # [vector size N, fraction descending for that vector size] 
        fraction_descending = []
        
        for i in range(0, len(all_rand_directions)):          
            w = np.array(start_weights[i])
            directions = all_rand_directions[i]
            possible_weights = w + alpha*directions

            # Check for steepest descent direction
            possible_costs = []
            for weight in possible_weights:
                possible_costs.append(g(weight))
            possible_costs = np.array(possible_costs)

            num_descending = 0
            for j in range(0, len(possible_weights)):
                if g(possible_weights[j]) < g(w):
                    num_descending += 1
            fraction_descending.append([i+1, num_descending/len(directions)])
            # [vector size N, fraction descending for that vector size]
    return fraction_descending
```

## Exercise 3: Coordinate Search and Coordinate Descent

```python
def coordinate_search(g, alpha_choice, max_its, w):
    """
    Looks for descent among the N coordinate directions and steps in the steepest one.,
    until a minimum of g is found.
    """


    # Generate coordinate directions that are normalized for the coordinate search
    normalized_coord_directions = np.zeros((2 * np.size(w), np.size(w)))  # 2N
    for i in range(0, np.size(w)):
        normalized_coord_directions[i, i] = 1
        normalized_coord_directions[i + np.size(w), i] = -1
    
    # Search the coordinates for steepest direction
    w_history = []  # container for weight history
    cost_history = []  # container for corresponding cost function history
    alpha = 0
    
    for k in range(1, max_its + 1):
    
        # Check if diminishing steplength rule used
        if alpha_choice == 'diminishing':
            alpha = 1 / float(k)
        else:
            alpha = alpha_choice
        cost_history.append(g(w))
        w_history.append(w)
    
        # Calculate possible weights based on direction and check for steepest direction
        possible_weights = w + alpha * normalized_coord_directions
        possible_costs = []
        for weight in possible_weights:
            possible_costs.append(g(weight))
        possible_costs = np.array(possible_costs)
        
        # Check for steepest descent direction
        i = np.argmin(possible_costs)
    
        # Step towards steepest descent direction
        if g(possible_weights[i]) < g(w):
            w = w + alpha * normalized_coord_directions[i, :]
    
    cost_history.append(g(w))
    w_history.append(w)

    return w_history, cost_history
```

```python
def coordinate_descent(g, alpha_choice, max_its, w):
    """
    Randomly go through the coordinate directions (use np.random.permutation)
    and the first time we find a descent, we use it.
    """

    # Generate coordinate directions that are normalized for the coordinate search
    N = np.size(w)  # The number of directions to test
    norm_directions = np.zeros((N, N))
    np.fill_diagonal(norm_directions, 1)

    # Search the coordinates for steepest direction
    w_history = []  # container for weight history
    cost_history = []  # container for corresponding cost function history
    alpha = 0

    for k in range(1, max_its + 1):

        # Check if diminishing steplength rule used
        if alpha_choice == 'diminishing':
            alpha = 1 / float(k)
        else:
            alpha = alpha_choice
        
        # Randomize the coordinate order
        norm_directions = np.zeros((N, N))
        np.fill_diagonal(norm_directions, 1)
        c = np.random.permutation(N)
        
        # Generate Random Directions and try N steps
        for j in range(0, N):
            test_direction = norm_directions[j]
            g_plus = g(w + alpha * test_direction)
            g_minus = g(w - alpha * test_direction)

            if g_plus < g_minus:
                w_new = w + alpha * test_direction
            elif g_minus < g_plus:
                w_new = w - alpha * test_direction
            
            # Take step if descending
            if g(w_new) < g(w):
                w = w_new
                cost_history.append(g(w_new))
                w_history.append(w)
                
 
    cost_history.append(g(w))
    w_history.append(w)

    return w_history, cost_history
```

## Exercise 4: Gradient Descent

```python
# import statement for autograd wrapped numpy
import autograd.numpy as np   
# import statment for gradient calculator
from autograd import grad  
import matplotlib.pyplot as pyplot 
```

Then we can use the following generic code for doing the gradient descent:

```python
# import automatic differentiator to compute gradient module
from autograd import grad 

# gradient descent function - inputs: g (input function), alpha (steplength parameter), max_its (maximum number of iterations), w (initialization)
def gradient_descent(g,alpha,max_its,w):
    # compute gradient module using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w]           # container for weight history
    cost_history = [g(w)]          # container for corresponding cost function history
    for k in range(max_its):
        # evaluate the gradient, store current weights and cost function value
        grad_eval = gradient(w)

        # take gradient descent step
        w = w - alpha*grad_eval
        
        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))
    return cost_history
```

We now want to find the minimum of 
$$g(w)=\frac{1}{50}(w^4+w^2+10w)$$ using the above gradient descent. Make three separate runs using a step length of $\alpha$=1,0.01 and 0.01 starting from $w^0=2$. Plot the resulting cost histories. Which step length works best for this particular function and initial point ?

```python
g = lambda w: (1/50)*(w**4 + w**2 + 10*w)

# Alpha = 1
x1 = gradient_descent(g, 1, 1000, np.array([2]))

# Alpha = 0.1
x01 = gradient_descent(g, 0.1, 1000, np.array([2]))

# Alpha = 0.01
x001 = gradient_descent(g, 0.01, 1000, np.array([2]))


pyplot.plot(np.linspace(0,len(x1),len(x1)), x1, color='r', label = 'alpha = 1')
pyplot.plot(np.linspace(0,len(x01),len(x01)), x01, color='b', label = 'alpha = 0.1')
pyplot.plot(np.linspace(0,len(x001),len(x001)), x001, color='g', label = 'alpha = 0.01')
pyplot.xlabel('k')
pyplot.ylabel('g(w^k)')
pyplot.title('Cost Function History Using Gradient Descent')
pyplot.legend(loc = "upper right")
```
