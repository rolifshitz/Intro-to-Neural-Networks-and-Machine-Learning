## Exercise 1: Momentum Accelerated Gradient Descent

### 1.1

```python
# gradient descent function - inputs: g (input function), 
# alpha (steplength parameter), 
# max_its (maximum number of iterations), w (initialization)
def momentum_accelerated_gd(g,alpha_choice,max_its,w,beta):
    # compute gradient module using autograd
    gradient = value_and_grad(g)
    # initialize gradient container
    N = w.shape
    d = np.zeros(N)
    alpha = 0
    
    # initialize the weight history and cost history containers
    weight_history = []
    cost_history = []
    cost_eval,grad_eval = gradient(w)
    
    for k in range(max_its):
        # Evaluate and store gradient direction, weights, and cost
        cost_eval,grad_eval = gradient(w)
        weight_history.append(w)
        cost_history.append(cost_eval)
        
        if alpha_choice == 'diminishing':
            alpha = 1/float(k)
        else:
            alpha = alpha_choice

        # calculate next gradient average
        d = beta*d + (1 - beta)*(-grad_eval)

        # take a step in the exponentially averaged gradient descent direction
        w = w + alpha * d
        
    # record weight and cost
    weight_history.append(w)
    cost_history.append(g(w))
    return weight_history,cost_history
```

```python
C = np.array([[0.5,0],[0,9.75]])
g = lambda w: (np.dot(np.dot(w.T,C),w))
w_history_0, c_history_0 = momentum_accelerated_gd(g,0.1,25,np.array([10, 1]),0)
w_history_01, c_history_01 = momentum_accelerated_gd(g,0.1,25,np.array([10, 1]),0.1)
w_history_07, c_history_07 = momentum_accelerated_gd(g,0.1,25,np.array([10, 1]),0.7)
```

## Exercise 2: Slow-Crawling Behavior of Gradient Descent

```python
def normalized_gd(g,alpha_choice,max_its,w, epsilon):
    # compute gradient module using autograd
    gradient = value_and_grad(g)
    
    # initialize the weight history and cost history containers
    weight_history = []
    cost_history = []
    alpha = 0
    
    for k in range(max_its):
        # Evaluate and store gradient direction, weights, and cost
        cost_eval,grad_eval = gradient(w)
        weight_history.append(w)
        cost_history.append(cost_eval)
        
        if alpha_choice == 'diminishing':
            alpha = 1/float(k)
        else:
            alpha = alpha_choice

        # normalize the gradient 
        norm = np.linalg.norm(grad_eval)
    
        if norm == 0:
            norm += epsilon
        grad_eval_norm = grad_eval/norm
    
        # step
        w = w - alpha*grad_eval_norm

    # record weight and cost
    weight_history.append(w)
    cost_history.append(g(w))
    return weight_history,cost_history
```

```python
# gradient descent function - inputs: g (input function), alpha (steplength parameter), max_its (maximum number of iterations), w (initialization)
def standard_gd(g,alpha,max_its,w):
    # compute gradient module using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w]           # container for weight history
    cost_history = [g(w)]          # container for corresponding cost function history
    for k in range(max_its):
        # evaluate the gradient, store current weights and cost function value
        grad_eval = gradient(w)

        # take gradient descent step
        w = w - alpha*grad_eval
        
        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history, cost_history
```

```python
from autograd import numpy as anp
g = lambda w: np.tanh(4*w[0] + 4*w[1]) + max(1, 0.4*w[0]**2) + 1

w_history_norm, c_history_norm = normalized_gd(g,0.1,1000,anp.array([2., 2.]),10**-7)
w_history_stand, c_history_stand = standard_gd(g,0.1,1000,np.array([2., 2.]))
pyplot.plot(np.linspace(0,len(c_history_norm),len(c_history_norm)), c_history_norm, color='r', label='Normalized GD')
pyplot.plot(np.linspace(0,len(c_history_stand),len(c_history_stand)), c_history_stand, color='b', label='Standard GD')

pyplot.xlabel('Iteration (k)')
pyplot.ylabel('Cost (g(w^k))')
pyplot.title('Cost Function History')
pyplot.legend(loc = "upper right")
```

**Comment on the observations, on the progress made with each approach:** <br>As can be seen in the figure above, the Standard GD algorithm was not able to make almost any progress (stopping the minimization at around cost = 3) while the Normalized GD was able to do so and minimize the cost function (stopping at around cost = 1). The cost function that is being minimized is very flat around the minimum, and, as such, it is difficult for the standard GD to continue once it reaches gradients that are very small (close to 0); the algorithm stops prior to reaching the minimum. However, using the Standard GD method, the gradient is normalized at each step such that the length of the step that is taken is not decreasing, but is a constant step length alpha. This allows the function to make more progress, even when reaching very flat regions and to eventually reach the minimum. For this reason, we see that the Standard GD algorithm stopped minimizing at around cost = 3, while the Normalized GD continued further and stopped minimizing at around cost = 1.

# Exercise 3: Student Debt - Linear Regression Using a Single Newton Step

### 3.1, 3.2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot 

# import the dataset
csvname = 'student_debt_data.csv'
data = np.asarray(pd.read_csv(csvname,header = None))
```

```python
# extract input
x = data[:,0]
x.shape = (len(x),1)

# pad input with ones 
o = np.ones((len(x),1))
x_new = np.concatenate((o,x), axis = 1)

# # extract input
x = data[:,0]
y = data[:,1]
```

```python
def newtons_method(x_new,y):
    A = np.dot(x_new.T,x_new)
    b = np.dot(x_new.T,y)
    w = np.dot(np.linalg.pinv(A), b)
    return w[0], w[1]
```

```python
bias, weight = newtons_method(x_new,y)

# Approximate linear regression
y_pred = []
for i in range(0,len(x)):
    y_pred.append(bias + weight*x[i])
y_pred = np.array(y_pred)

# Visualize the data
pyplot.scatter(x, y, color='r')
pyplot.plot(x, y_pred, color='b')
pyplot.xlabel("Year")
pyplot.ylabel("US Student Debt ($ Trillion)")
pyplot.title("US Student Debt from 2004-2014 (student_debt_data.csv)")
```

## Exercise 4: Linear regression with optimization

### 4.1, 4.2

```python
import autograd.numpy as np
from autograd import grad

# gradient descent function - inputs: g (input function), 
# alpha (steplength parameter), 
# max_its (maximum number of iterations), w (initialization)
def gradient_descent(g,alpha,max_its,w):
    # compute gradient module using autograd
    gradient = grad(g)

    # run the gradient descent loop
    weight_history = [w]     # container for weight history
    cost_history = [g(w)]    # container for corresponding cost function history
    for k in range(max_its):
        # evaluate the gradient, store current weights and cost function value
        grad_eval = gradient(w)

        # take gradient descent step
        w = w - alpha*grad_eval

        # record weight and cost
        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history,cost_history
```

```python
# Import the dataset
csvname = 'kleibers_law_data.csv'
data = np.loadtxt(csvname,delimiter=',')
x = data[:-1,:]
x = x[0]
y = data[-1:,:] 
y = y[0]

# Redefine dataset in terms of log(x) and log(y)
x_log = np.log(x)
y_log = np.log(y)
```

```python
# Define the linear model
def model(w, x):
    return (w[0] + x*w[1])
```

```python
# Define the cost function
def least_squares(w, x=x_log, y=y_log): # Returns the LSE cost.
    P = x.shape[0]
    LSE = np.sum((model(w,x) - y)**2)
    return (1/float(P))*LSE
```

```python
# Optimize (find the minimum) using gradient descent
w_history, c_history = gradient_descent(least_squares, 0.01, 1000, np.array([0.5,0.5]))
print("The [bias, weight] that minimizes the cost function is:")
w_history[1000]
```

### 4.3

```python
# Get the bias and weight for the linear regression model from the optimization
bias = w_history[1000][0]
weight = w_history[1000][1]

# Generate predictions using linear regression
y_pred = []
for i in range(0,len(x_log)):
    y_pred.append(bias + weight*x_log[i])
y_pred = np.array(y_pred)

# Visualize the data and plot linear regression
pyplot.scatter(x_log, y_log, color='r')
pyplot.plot(x_log, y_pred, color='b')
pyplot.xlabel("log(x); x is the body mass in kg")
pyplot.ylabel("log(y); y is the metabolic rate in KJ/day")
pyplot.title("Daily Energy Required by Animals Per Day (kleibers_law_data.csv)")
```

```python
pyplot.plot(np.linspace(0,len(c_history),len(c_history)), c_history, color='r')
pyplot.xlabel('Iterations (k)')
pyplot.ylabel('Cost (g(w^k))')
pyplot.title("Cost Function History for Kleiber's Law Optimization with Gradient Descent")
```
